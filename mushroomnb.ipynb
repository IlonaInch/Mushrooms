```python
df["class"] = df["class"].map({"p":False,"e":True})
```

**To predict we won't drop other columns because they are probably important and will be related to the target variable 'class'.**

```python
y = df["class"]
```

```python
X = df.drop(["class"], axis=1)
```

```python
X = pd.get_dummies(X, X.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

**The first model is Random Forest**

```python
clfRFC = RandomForestClassifier()
scoresRFC = cross_val_score(clfRFC, X, y, cv=15)

print("Random Forest Classifier's prediction accuracy is:", scoresRFC.mean())
```

**The second model is Gradient Boosting**

```python
clfGBF = GradientBoostingClassifier()
scoresGBF = cross_val_score(clfGBF, X, y, cv=15)

print("Gradient Boosting's prediction accuracy is:", scoresGBF.mean())
```

**The third model is Logistic Regression**

```python
clfLR = LogisticRegression()
scoresLR = cross_val_score(clfLR, X, y, cv=15)

print("Logistic Regression's prediction accuracy is:", scoresLR.mean())
```

```python
results = pd.DataFrame({
    'Model': ['Random Forest', ' Gradient Boosting', 
              'Logistic Regression'],
    'Score': [scoresRFC.mean(), scoresGBF.mean(), scoresLR.mean()]})

results
```

**We  see that all models could achieve very excellent accuracy.**
